```python
# ==========================================
# ✅ FULL CREWAI + HUGGINGFACE + FAISS + TAVILY SETUP (NO OPENAI)
# ==========================================

# --- Imports ---
import os
from transformers import pipeline
import torch
from crewai import Agent, Task, Crew, BaseLLM
from crewai_tools import DirectoryReadTool, WebsiteSearchTool
from tavily import TavilyClient
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings

# --- Suppress tokenizer warnings ---
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# ==========================================
# 1️⃣  LOAD LOCAL KNOWLEDGE BASE (FAISS)
# ==========================================
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vectorstore = FAISS.load_local(
    folder_path="faiss_index",
    embeddings=embeddings,
    allow_dangerous_deserialization=True
)

retriever = vectorstore.as_retriever()
print("✅ FAISS knowledge base loaded.")

# ==========================================
# 2️⃣  DEFINE LOCAL HUGGING FACE LLM CLASS
# ==========================================
class HFLocalLLM(BaseLLM):
    """Minimal BaseLLM subclass so CrewAI treats this as a valid LLM provider."""
    def __init__(self, model_name="google/flan-t5-base", max_new_tokens=256):
        self.generator = pipeline(
            "text2text-generation",
            model=model_name,
            max_new_tokens=max_new_tokens,
            device=-1  # CPU
        )

    def call(self, messages, **kwargs):
        prompt = "\n".join([m["content"] for m in messages])
        result = self.generator(prompt)[0]["generated_text"]
        return result

hf_llm = HFLocalLLM()
print("✅ Hugging Face LLM connected locally (flan-t5-base).")

# ==========================================
# 3️⃣  SET UP TOOLS (FAISS + TAVILY)
# ==========================================
TAVILY_API_KEY = "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"  # your existing key
tavily_client = TavilyClient(api_key=TAVILY_API_KEY)

document_tool = DirectoryReadTool(
    name="document_search_tool",
    directory_path="/Users/macbook/Desktop/internship/rag_project/crew_data",
    retriever=retriever
)

web_tool = WebsiteSearchTool(
    name="web_search_tool",
    client=tavily_client
)

# ==========================================
# 4️⃣  DEFINE AGENTS
# ==========================================
researcher = Agent(
    role="Researcher",
    goal="Find accurate and up-to-date information using FAISS and Tavily.",
    backstory="A data-driven researcher who finds precise and useful information from both local and online sources.",
    tools=[document_tool, web_tool],
    llm=hf_llm,
    verbose=True,
)

writer = Agent(
    role="Writer",
    goal="Write a clear, structured summary based on the researcher's findings.",
    backstory="A professional technical writer skilled at summarizing AI and data topics clearly.",
    tools=[document_tool],
    llm=hf_llm,
    verbose=True,
)

critic = Agent(
    role="Reviewer",
    goal="Review and refine the final answer for clarity, factual accuracy, and structure.",
    backstory="An experienced editor ensuring the final answer is polished and accurate.",
    tools=[document_tool],
    llm=hf_llm,
    verbose=True,
)

# ==========================================
# 5️⃣  DEFINE TASKS + CREW
# ==========================================
research_task = Task(
    description="Research the latest advancements in large language models using FAISS and Tavily.",
    expected_output="A structured summary of recent LLM advancements.",
    agent=researcher,
)

write_task = Task(
    description="Turn the research findings into a concise and professional explanation.",
    expected_output="A clear and engaging report on LLM advancements.",
    agent=writer,
)

review_task = Task(
    description="Edit and finalize the response for readability and accuracy.",
    expected_output="Final polished summary of LLM advancements.",
    agent=critic,
)

crew = Crew(
    agents=[researcher, writer, critic],
    tasks=[research_task, write_task, review_task],
    verbose=True,
)
# --- CrewAI bug-bypass patch ---
import crewai.crew as crew_mod

def safe_calculate_usage_metrics(self):
    """Bypass token-usage tracking for custom local LLMs."""
    class DummyUsage:
        def __init__(self): self.data = {}
        def add_usage_metrics(self, x): pass
    return DummyUsage()

crew_mod.Crew.calculate_usage_metrics = safe_calculate_usage_metrics
# --- end patch ---

# ==========================================
# 6️⃣  RUN THE CREW
# ==========================================
query = "latest advancements in large language models (LLMs)"
final_result = crew.kickoff(inputs={"query": query})

print("\n=== ✅ FINAL OUTPUT ===\n")
print(str(final_result)[:2000])  # Limit length for readability

# ==========================================
# 7️⃣  SAVE OUTPUT AS TEXT + PDF
# ==========================================
# --- Save as text ---
with open("final_report.txt", "w", encoding="utf-8") as f:
    f.write("Query: " + query + "\n\n")
    f.write(str(final_result))
print("✅ Final report saved as final_report.txt")

# --- Save as PDF ---
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def save_as_pdf(text, filename="final_report.pdf"):
    c = canvas.Canvas(filename, pagesize=letter)
    width, height = letter
    y = height - 50
    for line in text.split("\n"):
        c.drawString(50, y, line[:100])
        y -= 15
        if y < 50:
            c.showPage()
            y = height - 50
    c.save()
    print(f"✅ PDF saved as {filename}")

save_as_pdf(str(final_result))
```
